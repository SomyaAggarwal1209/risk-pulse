```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def label_encode(data, columns):
    """
    Perform label encoding on specified columns of a dataframe.

    Args:
    data (pd.DataFrame): The input dataframe.
    columns (list): List of column names to label encode.

    Returns:
    pd.DataFrame: The dataframe with label encoded columns.
    """
    label_encoders = {}
    for column in columns:
        le = LabelEncoder()
        data[column] = le.fit_transform(data[column])
        label_encoders[column] = le
    return data

# Example usage:
# df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Blue'], 'Size': ['S', 'M', 'L', 'M']})
# df, encoders = label_encode(df, ['Color', 'Size'])
# print(df)
```

```python
import pandas as pd

def one_hot_encode(data, columns):
    """
    Perform one-hot encoding on specified columns of a dataframe.

    Args:
    data (pd.DataFrame): The input dataframe.
    columns (list): List of column names to one-hot encode.

    Returns:
    pd.DataFrame: The dataframe with one-hot encoded columns.
    """
    data = pd.get_dummies(data, columns=columns, drop_first=True)
    return data

# Example usage:
# df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Blue'], 'Size': ['S', 'M', 'L', 'M']})
# df = one_hot_encode(df, ['Color', 'Size'])
# print(df)
```

```python
import pandas as pd

def split_rows(data):
    """
    Split rows in a DataFrame based on the values in 'loan_repayment_periods' and 'recent_loan_approvals_rejections' columns.
    
    Args:
    data (pd.DataFrame): The input dataframe.
    
    Returns:
    pd.DataFrame: The transformed dataframe with rows split based on 'loan_repayment_periods' and 'recent_loan_approvals_rejections'.
    """
    # Initialize an empty list to hold the new rows
    new_rows = []
    
    # Iterate over each row in the DataFrame
    for idx, row in data.iterrows():
        # Split the 'loan_repayment_periods' and 'recent_loan_approvals_rejections' columns by comma
        repayment_periods = row['loan_repayment_periods'].split(', ')
        approvals_rejections = row['recent_loan_approvals_rejections'].split(', ')
        
        # Ensure both lists have the same length
        if len(repayment_periods) == len(approvals_rejections):
            # Create a new row for each pair of repayment_period and approval_rejection
            for repayment_period, approval_rejection in zip(repayment_periods, approvals_rejections):
                new_row = row.copy()
                new_row['loan_repayment_periods'] = repayment_period
                new_row['recent_loan_approvals_rejections'] = approval_rejection
                new_rows.append(new_row)
        else:
            raise ValueError("Mismatch in lengths of 'loan_repayment_periods' and 'recent_loan_approvals_rejections' columns")
    
    # Create a new DataFrame from the new rows
    new_data = pd.DataFrame(new_rows)
    
    return new_data

# Example usage:

```

```python
import pandas as pd

def convert_repayment_periods_to_int(data):
    """
    Remove the 'years' postfix from the 'loan_repayment_periods' column and convert it to an integer.
    
    Args:
    data (pd.DataFrame): The input dataframe.
    
    Returns:
    pd.DataFrame: The transformed dataframe with 'loan_repayment_periods' converted to integer.
    """
    # Remove 'years' and 'year' postfix and convert to int
    data['loan_repayment_periods'] = data['loan_repayment_periods'].str.replace(' years', '').str.replace(' year', '').astype(int)
    return data

# Example usage:
df = pd.DataFrame({
    'id': [1, 2, 3],
    'total_number_of_loans': [2, 3, 3],
    'current_loan_balances': [15000.0, 30000.0, 30000.0],
    'loan_repayment_periods': ['5 years', '10 year', '7 years'],
    'outstanding_loan_amounts': [5000.0, 10000.0, 10000.0],
    'recent_loan_approvals_rejections': ['Approved', 'Rejected', 'Approved']
})

df = convert_repayment_periods_to_int(df)
print(df)
```

```python
import pandas as pd

def merge_dataframes_columnwise(dataframes):
    """
    Merge a list of DataFrames with different column names into a single DataFrame column-wise.
    
    Args:
    dataframes (list of pd.DataFrame): The list of DataFrames to merge.
    
    Returns:
    pd.DataFrame: The merged DataFrame.
    """
    merged_df = pd.concat(dataframes, axis=1)
    return merged_df

# Example usage:
df1 = pd.DataFrame({
    'id1': [1, 2, 3],
    'value1': [10, 20, 30]
})

df2 = pd.DataFrame({
    'id2': [4, 5, 6],
    'value2': [40, 50, 60]
})

df3 = pd.DataFrame({
    'id3': [7, 8, 9],
    'value3': [70, 80, 90]
})

dataframes = [df1, df2, df3]
merged_df = merge_dataframes_columnwise(dataframes)

print(merged_df)
```
