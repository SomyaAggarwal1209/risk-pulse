```python
import pandas as pd

# Load the datasets
bse_data = pd.read_excel('./Daily Turnover, Market Capitalisation and Traded Volume at BSE .xlsx')
nse_data = pd.read_excel('./Daily Turnover, Market Capitalisation and Traded Volume at NSE.xlsx')
scb_agg_169 = pd.read_excel('./HBS Table No. 169 _ Scheduled Commercial Banks - Select Aggregates.xlsx')
scb_agg_207 = pd.read_excel('./HBS Table No. 207 _ Scheduled Commercial Banks - Select Aggregates.xlsx')
forex_turnover = pd.read_excel('./Turnover in Foreign Exchange Market - Daily.xlsx')

# Display the first few rows of each dataset
bse_head = bse_data.head()
nse_head = nse_data.head()
scb_agg_169_head = scb_agg_169.head()
scb_agg_207_head = scb_agg_207.head()
forex_turnover_head = forex_turnover.head()
```

BSE data preprocessing

```python
# Inspect the BSE dataset
bse_data.info()

# Check for missing values
missing_values_bse = bse_data.isnull().sum()

# Convert the 'Date' column to datetime
bse_data['Date'] = pd.to_datetime(bse_data['Date'], errors='coerce')

# Drop rows with missing 'Date' values
bse_data = bse_data.dropna(subset=['Date'])

# Forward fill missing values for other columns
bse_data = bse_data.fillna(method='ffill')

# Display the cleaned BSE dataset info
bse_data.info()
missing_values_bse, bse_data.head()
```

NSE data preprocessing

```python
nse_data
```

```python
# Inspect the NSE dataset
nse_data.info()

# Check for missing values
missing_values_nse = nse_data.isnull().sum()

# Convert the 'Date' column to datetime
nse_data['Date'] = pd.to_datetime(nse_data['Date'], errors='coerce')

# Drop rows with missing 'Date' values
nse_data = nse_data.dropna(subset=['Date'])

# Convert numerical columns to appropriate types
nse_data.iloc[:, 1:] = nse_data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Forward fill missing values for other columns
nse_data = nse_data.fillna(method='ffill')

# Display the cleaned NSE dataset info
cleaned_nse_info = nse_data.info()
missing_values_nse, cleaned_nse_info, nse_data.head()
```

Merge dataset

```python
# Rename the columns for consistency
scb_agg_169.rename(columns={'Month\n': 'Date'}, inplace=True)
scb_agg_207.rename(columns={'Fortnight ended': 'Date'}, inplace=True)

# Ensure all datasets have a consistent datetime format for merging
bse_data['Date'] = pd.to_datetime(bse_data['Date'], errors='coerce')
nse_data['Date'] = pd.to_datetime(nse_data['Date'], errors='coerce')
scb_agg_169['Date'] = pd.to_datetime(scb_agg_169['Date'], errors='coerce')
scb_agg_207['Date'] = pd.to_datetime(scb_agg_207['Date'], errors='coerce')
forex_turnover['Date'] = pd.to_datetime(forex_turnover['Date'], errors='coerce')

# Merge BSE and NSE data on the Date column
merged_data = pd.merge(bse_data, nse_data, on='Date', suffixes=('_BSE', '_NSE'))

# Merge the bank aggregates and forex data on the Date column
merged_data = pd.merge(merged_data, forex_turnover, on='Date', how='left')
merged_data = pd.merge(merged_data, scb_agg_169, on='Date', how='left')
merged_data = pd.merge(merged_data, scb_agg_207, on='Date', how='left')

# Display the first few rows of the merged dataset
merged_data.head()
```

Feature engineering and Label creation

```python
# Calculate percentage change in market capitalization for BSE and NSE
merged_data['Market Cap Change_BSE'] = merged_data['Market Capitalisation (Rs. Crores)_BSE'].pct_change() * 100
merged_data['Market Cap Change_NSE'] = merged_data['Market Capitalisation (Rs. Crores)_NSE'].pct_change() * 100

# Define a threshold for market risk: Letting a drop of more than 2% indicates risk
risk_threshold = -2

# Create target variable based on market capitalization change
merged_data['Market Risk'] = ((merged_data['Market Cap Change_BSE'] < risk_threshold) | (merged_data['Market Cap Change_NSE'] < risk_threshold)).astype(int)

# Drop rows with missing values resulting from percentage change calculation
merged_data = merged_data.dropna()

# Display the first few rows of the dataset with the new target variable
merged_data.head()
```

```python
merged_data.shape
```

Feature Selection

```python
# Define the features and the target
features = merged_data.drop(columns=['Date', 'Market Risk'])
target = merged_data['Market Risk']

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a Random Forest classifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluate the model
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
classification_report = classification_report(y_test, y_pred)
confusion_matrix = confusion_matrix(y_test, y_pred)

classification_report, confusion_matrix
```

Cost-Sensitive Model Metrics (Final)

```python
# Define a Random Forest classifier with class_weight parameter
model_cost_sensitive = RandomForestClassifier(random_state=42, class_weight={0: 1, 1: 10})
model_cost_sensitive.fit(X_train, y_train)

# Evaluate the cost-sensitive model
y_pred_cost_sensitive = model_cost_sensitive.predict(X_test)
cost_sensitive_classification_report = classification_report(y_test, y_pred_cost_sensitive)
cost_sensitive_confusion_matrix = confusion_matrix(y_test, y_pred_cost_sensitive)

print(cost_sensitive_classification_report)
print(cost_sensitive_confusion_matrix)
```

Accuracy

```python
from sklearn.metrics import accuracy_score

# Cost-Sensitive Model Accuracy
cost_sensitive_accuracy = accuracy_score(y_test, y_pred_cost_sensitive)
print("Cost-Sensitive Model Accuracy:", cost_sensitive_accuracy)
```
